```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("international_matches.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
fifa_rankings = df[['date','home_team','away_team','home_team_fifa_rank','away_team_fifa_rank','home_team_goalkeeper_score', 'away_team_goalkeeper_score',
       'home_team_mean_defense_score', 'home_team_mean_offense_score',
       'home_team_mean_midfield_score', 'away_team_mean_defense_score',
       'away_team_mean_offense_score', 'away_team_mean_midfield_score']]
```

```python
home = fifa_rankings[['date','home_team','home_team_fifa_rank','home_team_goalkeeper_score','home_team_mean_defense_score','home_team_mean_offense_score','home_team_mean_midfield_score']].rename(columns={'home_team':'Team','home_team_fifa_rank':'Rank','home_team_goalkeeper_score':'GK','home_team_mean_defense_score':'DEF','home_team_mean_offense_score':'ATT','home_team_mean_midfield_score':'MID'})
```

```python
home.tail()
```

```python
away = fifa_rankings[['date','away_team','away_team_fifa_rank','away_team_goalkeeper_score','away_team_mean_defense_score','away_team_mean_offense_score','away_team_mean_midfield_score']].rename(columns={'away_team':'Team','away_team_fifa_rank':'Rank','away_team_goalkeeper_score':'GK','away_team_mean_defense_score':'DEF','away_team_mean_offense_score':'ATT','away_team_mean_midfield_score':'MID'})
```

```python
away.tail()
```

```python
fifa_rankings = pd.concat([home,away])
```

```python
fifa_rankings.tail()
```

```python
#Select each country latest match
fifa_rankings = fifa_rankings.sort_values(['Team','date'],ascending=[True,False])
fifa_rank = fifa_rankings.groupby('Team').first()
```

```python
fifa_rank = fifa_rank.reset_index()
fifa_rank.head()
```

```python
#Filter by teams in Worldcup
list_2022 = ['Qatar', 'Germany', 'Denmark', 'Brazil', 'France', 'Belgium', 'Croatia', 'Spain', 'Serbia', 'England', 'Switzerland', 'Netherlands', 'Argentina', 'IR Iran', 'Korea Republic', 'Japan', 'Saudi Arabia', 'Ecuador', 'Uruguay', 'Canada', 'Ghana', 'Senegal', 'Portugal', 'Poland', 'Tunisia', 'Morocco', 'Cameroon', 'USA', 'Mexico', 'Wales', 'Australia', 'Costa Rica']
final_df = fifa_rank[(fifa_rank["Team"].apply(lambda x: x in list_2022))]
```

```python
final_df.isnull().sum()
```

```python
#Qatar doesnt have any FIFA ratings as its host country and doesnt have played many matches
final_df[final_df['GK'].isnull()]
```

```python
#Giving average 60 for all Qatar ratings attributes
final_df = final_df.fillna(60)
```

```python
final_df.sort_values('Rank',inplace=True)
```

```python
final_df.drop('date',axis=1,inplace=True)
```

```python
final_df
```

```python
final_df.to_csv("FifaRankings.csv",index=False)
```
